```python
# Import our usual packages...
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Import some useful packages from scikit-learn
from sklearn.feature_extraction.text import CountVectorizer # transfer number into text data
from sklearn.metrics import *
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB # The Naive Bayes algo
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
```

```python
# replace the numerical price into the price category (20-33.3 as low, 33.3-46.6 as mid, 46.6-60 as high)

game_df['Price'] = np.where(
    (game_df['Price'] <= 33.3) & (game_df['Price'] >= 20),
    'Low',
    np.where(
        (game_df['Price'] > 33.3) & (game_df['Price'] <= 46.6),
        'Mid',
        np.where(
            (game_df['Price'] > 46.6) & (game_df['Price'] <= 60),
            'High',
            'Other/Out_of_Range'
        )
    )
)

# replace the user rating into the range of  category (10-23.3 as low, 23.3-36.6 as mid, 36.6-50 as high)

game_df['User Rating'] = np.where(
    (game_df['User Rating'] <= 23.3) & (game_df['User Rating'] >= 10),
    'Low',
    np.where(
        (game_df['User Rating'] > 23.3) & (game_df['User Rating'] <= 36.6),
        'Mid',
        np.where(
            (game_df['User Rating'] > 36.6) & (game_df['User Rating'] <= 50),
            'High',
            'Other/Out_of_Range'
        )
    )
)

# replace the game length into the length category (5-23.3 as short, 23.3-41.6.6 as mid, 41.6-60 as long)

game_df['Game Length (Hours)'] = np.where(
    (game_df['Game Length (Hours)'] <= 23.3) & (game_df['Game Length (Hours)'] >= 5),
    'Short',
    np.where(
        (game_df['Game Length (Hours)'] > 23.3) & (game_df['Game Length (Hours)'] <= 41.6),
        'Mid',
        np.where(
            (game_df['Game Length (Hours)'] > 41.6) & (game_df['Game Length (Hours)'] <= 60),
            'Long',
            'Other/Out_of_Range'
        )
    )
)

# replace the numerical min number players into the price category (20-33.3 as low, 33.3-46.6 as mid, 46.6-60 as high)

game_df['Min Number of Players'] = np.where(
    (game_df['Min Number of Players'] <= 3) & (game_df['Min Number of Players'] >= 1),
    'Small',
    np.where(
        (game_df['Min Number of Players'] > 3) & (game_df['Min Number of Players'] <= 6),
        'Mid',
        np.where(
            (game_df['Min Number of Players'] > 6) & (game_df['Min Number of Players'] <= 10),
            'Large',
            'Other/Out_of_Range'
        )
    )
)


print(game_df.info())
```

```python
# drop game title and release year columns
game_df = game_df.drop(columns=['Game Title', 'Release Year', 'Developer', 'Publisher', 'User Review Text'])

print(game_df.info())
game_df.head()
```

```python


# Define inputs and target
# X_encoded will include all features except the target 'Price'
X_encoded = game_df.drop(columns=['Price'])

# Perform one-hot encoding on all categorical features in X_encoded
X_encoded = pd.get_dummies(X_encoded)

# y is the target variable 'Price'
y_target = game_df['Price']
```

```python
# split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_encoded,y_target , test_size=0.3, random_state=0)
```

## Drumroll.... Time to run the Naive Bayes Classifier

```python
# Let's call the model 'NB' for short
NB = MultinomialNB()
NB.fit(X_train, y_train)
```

```python
# How many entries were ham vs spam?
np.set_printoptions(suppress=True, precision=0, formatter={'float_kind': '{:g}'.format})
NB.class_count_
```

```python
# Accuracy of NB model
y_pred = NB.predict(X_test)
accuracy_score(y_test, y_pred)
```
