# Gradient Boosting

Fits two gradient boosting variants on the price prediction task:

- `HistGradientBoostingRegressor` (histogram-binned, fastest)
- `GradientBoostingRegressor` (classic sklearn)

Both use early-tuned hyperparameters (lr=0.05, max_depth ≈ 5–8, 300 trees).

```python
import sys, time
sys.path.insert(0, '..')
import numpy as np, pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor, HistGradientBoostingRegressor
from sklearn.metrics import r2_score, mean_squared_error
from src.preprocessing import build_features

X, y, _ = build_features('../data/video_game_reviews.csv')
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=67)
print(Xtr.shape, Xte.shape)
```

## HistGradientBoostingRegressor

```python
t0 = time.perf_counter()
hgb = HistGradientBoostingRegressor(max_iter=300, learning_rate=0.05, max_depth=8, random_state=0)
hgb.fit(Xtr, ytr)
elapsed = time.perf_counter() - t0
pred = hgb.predict(Xte)
print(f'R^2 = {r2_score(yte, pred):.4f}')
print(f'RMSE = {np.sqrt(mean_squared_error(yte, pred)):.4f}')
print(f'train time = {elapsed:.2f}s')
```

## GradientBoostingRegressor (sklearn)

```python
t0 = time.perf_counter()
gbm = GradientBoostingRegressor(n_estimators=300, learning_rate=0.05, max_depth=5, random_state=0)
gbm.fit(Xtr, ytr)
elapsed = time.perf_counter() - t0
pred = gbm.predict(Xte)
print(f'R^2 = {r2_score(yte, pred):.4f}')
print(f'RMSE = {np.sqrt(mean_squared_error(yte, pred)):.4f}')
print(f'train time = {elapsed:.2f}s')
```

## Feature importance (sklearn GBM)

```python
import matplotlib.pyplot as plt
imp = pd.Series(gbm.feature_importances_, index=Xtr.columns).sort_values(ascending=True)
ax = imp.tail(15).plot(kind='barh', figsize=(8, 6))
ax.set_xlabel('Importance')
ax.set_title('Top 15 GBM Feature Importances')
plt.tight_layout()
plt.show()
```
